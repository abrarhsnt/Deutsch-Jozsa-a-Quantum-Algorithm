```python
from qiskit import execute
provider = IBMQ.get_provider(hub="ibm-q-education", group="brac-uni-1",project="cse481-quantum")
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
##Deutsch-Jozsa Algortihm Balance
qc = QuantumCircuit(4,4)
qc.h([0,1,2])
qc.x(0)
qc.x(2)
qc.x(3)
qc.h(3)
qc.barrier()
qc.cx(0,3)
qc.cx(1,3)
qc.cx(2,3)
qc.barrier()
qc.x(0)
qc.x(2)
qc.h([0,1,2])
qc.barrier()
qc.measure([0,1,2],[0,1,2])
qc.draw()
```

```python
simul = provider.get_backend('ibmq_qasm_simulator')
job = execute(qc, backend=simul, shots=1200)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)

```
